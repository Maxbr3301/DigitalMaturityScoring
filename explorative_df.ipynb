```python
import os
import PyPDF2
import spacy
import pandas as pd
from collections import Counter

# Definieren Sie den Ordnerpfad mit Ihren PDFs
folder_path = '/Users/maxbreitruck/Library/CloudStorage/OneDrive-Universita虉tSt.Gallen/001_IWI/Dissertation/Paper#2/Literatur'

# Laden Sie ein NLP-Modell (z. B. spaCy) f眉r Textverarbeitung
nlp = spacy.load("de_core_news_sm")

# Verwenden Sie ein leeres W枚rterbuch, um Praktiken und Schl眉sselw枚rter zu speichern
practices_keywords = {
    "Praktik1": ["Keyword1", "Keyword2"],
    "Praktik2": ["Keyword3", "Keyword4"]
}

# Verwenden Sie eine Liste, um Daten f眉r den DataFrame zu speichern
data_for_df = []

# Durchlaufen Sie PDF-Dateien im angegebenen Ordner
for filename in os.listdir(folder_path):
    if filename.endswith('.pdf'):
        file_path = os.path.join(folder_path, filename)
        
        # PDF-Text extrahieren
        pdf_text = ""
        with open(file_path, 'rb') as pdf_file:
            pdf_reader = PyPDF2.PdfReader(pdf_file)
            for page_num in range(len(pdf_reader.pages)):
                page = pdf_reader.pages[page_num]
                pdf_text += page.extract_text()
        
        # Textverarbeitung mit spaCy
        doc = nlp(pdf_text)
        
        # Suchen Sie nach Praktiken und Schl眉sselw枚rtern in Ihrem Text
        for practice, keywords in practices_keywords.items():
            practice_count = 0
            for keyword in keywords:
                keyword_count = pdf_text.lower().count(keyword.lower())
                practice_count += keyword_count
            
            if practice_count > 0:
                data_for_df.append({
                    "Filename": filename,
                    "Praktik": practice,
                    "Keyword_Count": practice_count
                })

# Erstellen Sie einen DataFrame
df = pd.DataFrame(data_for_df)

# Anzeigen des DataFrames
print(df)
```
