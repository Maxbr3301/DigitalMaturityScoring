```python
from PyPDF2 import PdfReader

def extract_text_from_pdf(file_path):
    with open(file_path, 'rb') as file:
        reader = PdfReader(file)
        text = ""
        for page in reader.pages:
            text += page.extract_text()
    return text
```

```python
def search_keywords(text, keywords):
    counts = {keyword: len(re.findall(re.escape(keyword), text, re.IGNORECASE)) for keyword in keywords}
    return counts
```

```python
folder_path = '/Users/maxbreitruck/Library/CloudStorage/OneDrive-UniversitaetSt.Gallen/001_IWI/Dissertation/Paper#2/Literatur' 
# Using a list to store data for DataFrame
data_for_df = []

for filename in os.listdir(folder_path):
    if filename.endswith('.pdf'):
        file_path = os.path.join(folder_path, filename)
        text = extract_text_from_pdf(file_path)

        for pillar, capabilities in pillars_data.items():
            for capability, practices in capabilities.items():
                for practice, keywords in practices.items():
                    counts = search_keywords(text, keywords)
                    for keyword, count in counts.items():
                        data_for_df.append({
                            "Filename": filename,
                            "Pillar": pillar,
                            "Capability": capability,
                            "Practice": practice,
                            "Keyword": keyword,
                            "Count": count
                        })

# Create DataFrame
df_results = pd.DataFrame(data_for_df)

# Display DataFrame
print(df_results)
```

```python
def extract_relevant_text(file_path):
    with open(file_path, 'rb') as file:
        reader = PyPDF2.PdfReader(file)
        full_text = ""
        for page in reader.pages:
            full_text += page.extract_text() + "\n"

    # Find the start and end of the relevant section
    start_idx = full_text.find('Abstract')
    end_idx = full_text.find('References')

    if start_idx != -1 and end_idx != -1:
        return full_text[start_idx:end_idx]
    else:
        # Return full text if specific sections are not found
        return full_text
```
